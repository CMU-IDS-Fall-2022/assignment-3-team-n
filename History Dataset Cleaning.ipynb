```python
df = df.replace({'Home Team Name': {'Germany FR': 'Germany'}})
df = df.replace({'Home Team Name': {'Germany DR': 'Germany'}})
df = df.replace({'Home Team Name': {'Korea Republic': 'South Korea'}})
df = df.replace({'Home Team Name': {'IR Iran': 'Iran'}})
df = df.replace({'Away Team Name': {'Germany FR': 'Germany'}})
df = df.replace({'Away Team Name': {'Germany DR': 'Germany'}})
df = df.replace({'Away Team Name': {'Korea Republic': 'South Korea'}})
df = df.replace({'Away Team Name': {'IR Iran': 'Iran'}})
```
